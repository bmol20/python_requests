```python
header = {'authorization': 'Bearer' + acess_token,
          'X-Github-Api-Version': '2022-11-28'}
```

```python
api_base_url = 'https://api.github.com'
owner = 'amzn'
url = f'{api_base_url}/users/{owner}/repos'
url
```

```python
repos_list = []
for page in range(1,7):
    try:
        url_page = f'{url}?page={page}'
        response = requests.get(url_page, headers=header)
        print(response.json())
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

```python
repos_list[2][0]
```

```python
len(repos_list)
```

```python
soma = 0
for teste in range(1,len(repos_list)):
    soma += len(repos_list[teste])
soma
```

```python
repos_names = []
```

```python
repos_names = []
for page in repos_list:
    for repo in page:
        repos_names.append(repo['name'])

print(repos_names)
```

```python
print(len(repos_names))
```

```python
repos_lang = []

for page in repos_list:
    for repo in page:
        repos_lang.append(repo['language'])

print(repos_lang)
```

```python
print(len(repos_lang))
```

```python
repos_names_lang = {}

for indice in range(0, len(repos_names)):
    try:    
        repos_names_lang[repos_names[indice]] = repos_lang[indice]
    except:
        repos_names_lang[repos_names[indice]] = None

print(repos_names_lang)
```

```python
api_base_url = 'https://api.github.com'
owner = 'amzn'
url = f'{api_base_url}/users/{owner}/repos'
```

```python
import re
```

```python
header_total_pages = {'autorization': 'Bearer' + acess_token,
                        'Accept': 'application/vnd.github.v3+json'}
response = requests.get(url, headers=header_total_pages)

link_header = response.headers.get('Link', '')

pages = 1

if 'rel="last"' in link_header:
    match = re.search(r'page=(\d+)>; rel="last"', link_header)
    if match:
        pages = int(match.group(1))

print(f"Total de páginas: {pages}")
```
